```python
import re

def change_to_cat_style(sentence):
    sentence.strip()
    if len(sentence) <2:
        print(sentence)
        return '에러'
    sentence=re.sub(r'다.$|요.$|죠.$|지.$',sentence[-2]+'냥.',sentence)
    sentence=re.sub(r'습니다$|습니까$|나요$',sentence[-2]+'냥?',sentence)
    sentence=re.sub(r'양|얌|어요.$|지만.','냥',sentence)
    sentence=re.sub(r'텐데.$','거냥.',sentence)
    sentence.strip()
    return sentence
```

```python
with open('kor_sent_0.txt','r',encoding='utf-8') as file:
    text=file.read()
    text=text.split('>')
    new=list(map(change_to_cat_style,text))
new
```

```python
from konlpy.tag import Okt
okt=Okt()
str='저는 곽태경이고 컴퓨터를 좋아합니다.'
str_split=str.split(' ')
okt.pos(str)
```

```python
str='저는 곽태경이고 컴퓨터를 좋아합니다.'
def add_space_and_pos(str):
    str_split=str.split(' ')
    final_str=[]
    for i in str_split:
        final_str=[*final_str,*okt.pos(i)]
        final_str+= [(' ','Space')]
    return final_str
```

```python
add_space_and_pos('안녕하세요? 저는 버스를 타다가 넘어졌습니다. ㅠㅠ')
```

```python
from konlpy.tag import Okt
from random import randint
random_num=randint(0,10)
def make_cat_style_by_pos(sentence):
    okt_pos=add_space_and_pos(sentence)
    char_location=0
    for pos in okt_pos:
        char_location+=len(pos[0])
        if pos[1]=='Adjective':
            if pos[0][-1]=='다':
                sentence=sentence[:char_location]+'냥'+sentence[char_location:]
                char_location+=1
        elif pos[1]=='Noun':
            if pos[0]=='내' or pos[0]=='나' or pos[0]=='제' or pos[0]=='저':
                sentence=sentence[:char_location-1]+'고양이'+sentence[char_location:]
                char_location+=2
            elif pos[0]=='요':
                sentence=sentence[:char_location-1]+'냥'+sentence[char_location:]
        elif pos[1]=='Josa':
            if pos[0][-1]=='다':
                sentence=sentence[:char_location]+'냥'+sentence[char_location:]
                char_location+=1
            elif pos[0]=='이에요':
                sentence=sentence[:char_location-3]+'이다냥'+sentence[char_location:]
            elif randint(0,10)>=5:
                sentence=sentence[:char_location]+', 냥,'+sentence[char_location:]
                char_location+=4
    return sentence
```

```python
make_cat_style_by_pos('한국 사람들과 저는 컴퓨터를 좋아할까요?')
```
